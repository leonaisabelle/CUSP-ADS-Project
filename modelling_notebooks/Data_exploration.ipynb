```python
boroughs = pd.read_csv('../data/NTA_Borough_Dict.csv', index_col=0)
boroughs.head()
```

```python
data_wBoro = data.merge(boroughs, left_index=True, right_index=True, validate='1:1')
data_wBoro.head()
```

```python
garden_counts = data_wBoro[['GardenCounts', 'Borough']][data_wBoro.GardenCounts > 0]
garden_counts.head()
```

```python
data_wBoro.GardenCounts.sum()
```

```python
import numpy as np 

counts = pd.pivot_table(garden_counts, values='GardenCounts', index='Borough', aggfunc=np.sum)
counts
```

```python
# number of NTAs with gardens 
data_wBoro.Borough[data_wBoro.HasGarden > 0].value_counts()
```
